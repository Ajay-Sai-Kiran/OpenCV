Changing Images Color Profiles

```python
import cv2
from google.colab.patches import cv2_imshow

image=cv2.imread("/content/crying-earth-wearing-mask-blue-sky_1639-12021.png")
cv2_imshow(image)
```

```python
print(image.shape)
```

```python
# Grayscale image nothing but black & white
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
print(gray.shape)
```

```python
#Purpose of conversion is to decrease the channel
```

```python
#HSV[HUE SATURATION VALUE] Image
HSV=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
print(HSV.shape)
cv2_imshow(HSV)
```
